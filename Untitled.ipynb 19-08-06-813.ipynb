```python
import requests
import re
from bs4 import BeautifulSoup
import string
import pandas as pd
def Ref(html_name): 
    Ref = []
    with open('/Users/langchen777/Desktop/New Directions for Evaluations/HTML/'+html_name+'.html', 'r', encoding = 'utf-8') as f:
        soup = BeautifulSoup(f.read())
        content = soup.findAll("li")
    for li in content:
        authors = li.findAll("span", class_="author")
        line = []
        for x in authors:
            line.append(x.contents[0])
        while(len(line)<5):
            line.append("")
        while(len(line)>5):
            line.pop()
        try:
            title = li.find("span", class_="articleTitle").text
        except AttributeError:
            pass
        try:
            title = li.find("span", class_="chapterTitle").text
        except AttributeError:
            pass
        try:
            title = li.find("span", class_="bookTitle").text
        except AttributeError:
            pass
        try:
            title = li.find("span", class_="otherTitle").text
        except AttributeError:
            pass
        line.append(title)
        Ref.append(line)
    return Ref
```

```python
def Issue(issue):
    df = pd.DataFrame(columns=col)
    file = open('/Users/langchen777/Desktop/New Directions for Evaluations/issues/'+issue+'.txt', 'r')
    Lines = file.readlines()
    count = 0
    for line in Lines:
        start = line[:2]
        if start == "TY":
            count = count + 1
            #Ref = Ref()
            row = []
            row.append(line.split("-",1)[1].strip())
        elif start == "AU":
            row.append(line.split("-",1)[1].strip())
        elif start == "TI":
            while(len(row)<11):
                row.append("")
            while(len(row)>11):
                row.pop()
            row.append(line.split("-",1)[1].strip())
        elif start in ["JO","VL","IS","SN","DO","SP","EP"]:
            row.append(line.split("-",1)[1].strip())
        elif start == "ER":
            try:
                for r in Ref(issue+str(count)):
                    combine = r + row
                    a_series = pd.Series(combine, index = col)
                    df = df.append(a_series, ignore_index=True)
            except FileNotFoundError:
                pass
            else:
                r = ["","","","","",""]
                combine = r + row
    return df
```
